```python
def get_sentiment(rev):
    """Function which return the sentiment of the passed review as positive, negative or Neutral"""
    analysis = TextBlob(rev)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity <0:
        return 'negative'
    else:
        return 'neutral'
```

```python
import random
from nltk.corpus import movie_reviews

def get_reviews(num_of_review, category):
    """Returns a list of randomly selected reviews and their sentiments."""

    # category should be 'pos' or 'neg'
    file_ids = movie_reviews.fileids(category)

    # Select random review files
    selected_reviews = random.sample(file_ids, num_of_review)

    print("First two reviews:\n")

    print(movie_reviews.raw(selected_reviews[0])[:500], "\n")

    if num_of_review > 1:
        print(movie_reviews.raw(selected_reviews[1])[:500], "\n")

    final_processed_review = []

    for file_id in selected_reviews:
        review_text = movie_reviews.raw(file_id)

        parsed_review = {
            "text": review_text,
            "sentiment": get_sentiment(review_text)
        }

        final_processed_review.append(parsed_review)

    return final_processed_review
```

```python

import random
import nltk
from nltk.corpus import movie_reviews
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('movie_reviews')
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

def get_sentiment(text):


    score = sia.polarity_scores(text)["compound"]

    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

def get_reviews(num_of_review, category):


    file_ids = movie_reviews.fileids(category)


    num_of_review = min(num_of_review, len(file_ids))

    selected_files = random.sample(file_ids, num_of_review)

    print("=" * 70)
    print("FIRST TWO SELECTED REVIEWS")
    print("=" * 70)

    print("\nReview 1:\n")
    print(movie_reviews.raw(selected_files[0])[:700])

    if num_of_review > 1:
        print("\n" + "=" * 70)
        print("Review 2:\n")
        print(movie_reviews.raw(selected_files[1])[:700])

    final_processed_review = []

    for file in selected_files:

        review_text = movie_reviews.raw(file)

        parsed_review = {
            "text": review_text,
            "sentiment": get_sentiment(review_text)
        }

        final_processed_review.append(parsed_review)

    return final_processed_review


def analyze():


    print("\nAvailable Categories")
    print(movie_reviews.categories())

    category = input("\nEnter category (pos/neg): ").lower()

    if category not in movie_reviews.categories():
        print("\nInvalid category.")
        return

    try:
        num_reviews = int(input("Enter number of reviews: "))
    except ValueError:
        print("Please enter a valid integer.")
        return

    reviews = get_reviews(num_reviews, category)

    positive = sum(1 for review in reviews if review["sentiment"] == "positive")
    negative = sum(1 for review in reviews if review["sentiment"] == "negative")
    neutral = sum(1 for review in reviews if review["sentiment"] == "neutral")

    print("\n" + "=" * 70)
    print("SENTIMENT ANALYSIS RESULT")
    print("=" * 70)

    print(f"Total Reviews      : {num_reviews}")
    print(f"Positive Reviews   : {positive}")
    print(f"Negative Reviews   : {negative}")
    print(f"Neutral Reviews    : {neutral}")

    print("\nPercentage Summary")
    print("-" * 70)

    print(f"Positive : {(positive/num_reviews)*100:.2f}%")
    print(f"Negative : {(negative/num_reviews)*100:.2f}%")
    print(f"Neutral  : {(neutral/num_reviews)*100:.2f}%")
analyze()
```
