# REGEX

## What is Regex (Regular Expressions)?
       Regex (short for Regular Expressions) is like a search pattern used to match text.

     1.It’s basically a sequence of characters that defines a rule for finding strings inside text.

     2.Think of it like a CTRL+F on steroids → much more powerful than normal search.

## Why Do We Need Regex?
-- In Machine Learning, Regex is not a model, but a preprocessing / feature engineering tool.We use it mostly in text-based ML tasks (NLP).

We need regex when we want to:

  1. Search → Find a pattern in text

  2. Validate → Check if input follows rules

  3. Extract → Pull out useful pieces of text

  4. Replace/Clean → Modify text based on patterns



```python
import pandas as pd  # Import the Pandas library for data manipulation

# Load the CSV file into a DataFrame
df = pd.read_csv("/content/reviews.csv")

# Display the first 5 rows of the DataFrame
df.head()
```

```python
import re  # Import the regular expressions module for text pattern matching
```

```python
# Display reviews that start with the word "love" (case-insensitive)
df[df['Review'].str.contains(r"^love", flags=re.IGNORECASE, regex=True)]
```

```python
# Display email addresses that start with "bob"
df[df['Email'].str.contains(r'^bob', regex=True)]
```

```python
# Display phone numbers that start with "9"
df[df['Phone'].astype(str).str.contains(r'^9', regex=True)]
```

```python
# Display names that start with a vowel
df[df['Name'].str.contains(r'^[AEIOUaeiou]', regex=True)]
```

```python
# Display email addresses that end with "example.com"
df[df['Email'].str.contains(r'example.com$', regex=True)]
```

```python
# Display phone numbers that end with "5"
df[df['Phone'].astype(str).str.contains(r'5$', regex=True, na=False)]
```

```python
# Replace '@' and '.' with '_' in the Email column
df['Email'] = df['Email'].str.replace(r'[@.]', '_', regex=True)
```
