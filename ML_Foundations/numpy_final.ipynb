```python
a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)
print(b)

print(a+b)
# 4,3 1,3
# 4,3 3,3
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3)

print(a)
print(b)

#print(a+b)
# (3,4 ) (1,3)
#(3,4) (3,3)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)

print(a+b)
# (1,3) (3,1)
# (3,3) (3,3)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)

print(a)
print(b)

print(a + b)
#(1,3) (4,1)
# (4,3) (4,3)
```
